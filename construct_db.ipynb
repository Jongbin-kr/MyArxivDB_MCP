```python
from server import add_new_project, assign_paper_to_specific_project
from datetime import date


import re

def extract_arxiv_title_mapping(bibliography: str) -> dict:
    """
    Extracts a dictionary mapping arXiv IDs to paper titles from a bibliography string.
    
    Parameters:
        bibliography (str): The bibliography text to process.

    Returns:
        dict: A dictionary with arXiv ID (str) as key and paper title (str) as value.
    """
    arxiv_dict = {}
    
    # 정규표현식 패턴 정의
    title_pattern = r'“(.+?)”'
    arxiv_pattern = r'arXiv[:\s]*((\d{4}\.\d{4,5})(v\d+)?)|https?://arxiv\.org/abs/(\d{4}\.\d{4,5})(v\d+)?'
    
    # 줄 단위로 처리
    for line in bibliography.split('\n'):
        title_match = re.search(title_pattern, line)
        if not title_match:
            continue
        title = title_match.group(1)
        
        arxiv_match = re.search(arxiv_pattern, line)
        if not arxiv_match:
            continue
        
        arxiv_id = arxiv_match.group(1) if arxiv_match.group(1) else arxiv_match.group(4)
        arxiv_dict[arxiv_id] = title
    
    return arxiv_dict
```

```python
import random
random.seed(42)  # For reproducibility


def make_shuffled_testset():
    # 프로젝트별 bibliography 문자열 정의
    projects = {
        "casual_inference": Casual_inference,
        "continual_PEFT": continual_PEFT,
        "deep_morphometric": deep_morphometric,
        "domain_shift": domain_shift,
        "VLM": VLM
    }
    
    # 전체 논문 리스트
    all_papers = []

    for label, bib in projects.items():
        paper_map = extract_arxiv_title_mapping(bib)
        for arxiv_id, title in paper_map.items():
            all_papers.append({
                "arxiv_id": arxiv_id,
                "title": title,
                "label": label
            })
    
    # 셔플
    random.shuffle(all_papers)
    return all_papers


testset = make_shuffled_testset()
for i, paper in enumerate(testset):
    print(f"assign {paper['arxiv_id']} paper to nearest project")
    print(f"[{i}] Label: {paper['label']}, Title: {paper['title']}")
    print("\n")
    
"""
domain shift VXV
continual PEFT XX
deeep morphometric XVV
VLM VVV
causal inference VVX

"""
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
import pandas as pd

# Model assignments (predictions)
model_assignments = {
    "Domain-shift continual learning": ["2102.05082", "2011.12328", "2108.07258", "2412.13860", "2104.09937"],
    "PEFT for LLM personalization": ["2308.04014", "2411.19128", "2403.19211", "2205.05124"],
    "VLM for compositonal reasoning": ["2208.00005", "2111.11432", "2305.14676", "2108.05921"],
    "Deep Morphometric Profiling": ["2102.11107", "2409.11654", "2411.02572", "2209.07819", "1912.00318"],
    "Causal Inference of Economic Policies": ["2402.15585", "2412.19024"],
}

# Ground truth
true_assignments = {
    "Domain-shift continual learning": {"2308.04014", "2412.13860", "2104.09937", "2102.05082"},
    "PEFT for LLM personalization": {"2403.19211", "2411.19128", "2011.12328", "2205.05124"},
    "VLM for compositonal reasoning": {"2208.00005", "2108.05921", "2111.11432", "2305.14676"},
    "Deep Morphometric Profiling": {"2209.07819", "1912.00318", "2409.11654", "2411.02572"},
    "Causal Inference of Economic Policies": {"2102.11107", "2412.19024", "2108.07258", "2402.15585"},
}

# Flattened ground truth for lookup
truth_lookup = {}
for label, ids in true_assignments.items():
    for i in ids:
        truth_lookup[i] = label

# Collect predictions and true labels
y_true, y_pred = [], []
for pred_label, ids in model_assignments.items():
    for arxiv_id in ids:
        y_pred.append(pred_label)
        y_true.append(truth_lookup.get(arxiv_id, "Unknown"))

# Evaluation
report = classification_report(y_true, y_pred, output_dict=True)
df_report = pd.DataFrame(report).transpose()
accuracy = accuracy_score(y_true, y_pred)
f1_macro = f1_score(y_true, y_pred, average='weighted')

print(report)
print("Accuracy:", accuracy)
print("F1 Score (weighted):", f1_macro)

df_report
```
